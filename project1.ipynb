Distribution of Playersâ€™ Overall Ratings

```python
plt.Figure(figsize=(8,5))
sns.histplot(data=df,x='Overall')
plt.xlabel('Overall Rting')
plt.ylabel('Number of players')
plt.show()
```

Age of the players

```python
plt.Figure(figsize=(8,5))
print(df['Age'].median())
sns.histplot(x='Age',data=df)
plt.xlabel("Age")
plt.ylabel("Number of players")
plt.show()
```

Top 10 Nationalities by Player Count

```python
plt.figure(figsize=(8,5))
top10=df["Nationality"].value_counts().head(10)
data = top10.reset_index()
data.columns=['Nationality','Player Count']
sns.barplot(x="Nationality",y="Player Count",data=data,color='Orange',edgecolor='black')
plt.title("Top 10 Nationalities by Player Count")
plt.xlabel('Nationality')
plt.ylabel("Number of players")
plt.show()
```

Age vs Overall Rating

```python
plt.figure(figsize=(8,5))
plt.scatter(df['Age'],df['Overall'],color='orange')
plt.xlabel('Age')
plt.ylabel("Overall Ratin")
plt.title("Age vs Overall Rating")
plt.show()
```

Strong Foot

```python
plt.figure(figsize=(8,5))
colors=['red','orange']
sns.countplot(data=df,x='Preferred Foot',hue="Preferred Foot",palette=colors)
plt.show()
```

Top 10 Clubs by Average Overall Rating

```python
plt.figure(figsize=(8,5))
topclubs= df.groupby('Club')['Overall'].mean().sort_values(ascending=False).head(10)
topclubs.plot(kind='barh', color='green')
plt.title('Top 10 Clubs by Avg Overall Rating')
plt.xlabel('Average Overall Rating')
```

Age for Top 10 Nationalities

```python
top10=df['Nationality'].value_counts().head(10).index
topnations = df[df['Nationality'].isin(top10)]
plt.figure(figsize=(12,6))
sns.boxplot(data=topnations, x='Nationality', y='Age',hue="Nationality",palette='Set3')
plt.title('Age for Top 10 Nationalities')
plt.xticks(rotation=45)
plt.xlabel('Nationality')
plt.ylabel('Age')
plt.show()
```
